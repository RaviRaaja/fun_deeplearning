```python
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(2)
dataset = np.loadtxt("diab.csv", delimiter=',')
X = dataset[:, 0:8]
Y = dataset[:, 8]
```

```python
def base_model(OPT):
    model = Sequential()
    model.add(Dense(13, input_dim=8, kernel_initializer='uniform',activation='relu'))
    model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
    model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=OPT, metrics=['accuracy'])
    return model
```

```python
opt = 'sgd'
model = base_model(opt)
history = model.fit(X, Y, validation_split=0.33, epochs=150,batch_size=1, verbose=1)
```

```python
#print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('MODEL ACCURACY')
plt.ylabel("ACCURACY")
plt.xlabel("EPOCH")
plt.legend(['train','test'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('MODEL LOSS')
plt.ylabel("LOSS")
plt.xlabel("EPOCH")
plt.legend(['train','test'], loc='upper left')
plt.show()
```
